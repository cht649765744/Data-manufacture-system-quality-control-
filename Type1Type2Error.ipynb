```python
#import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

```python
#Problem1.1
import numpy as np
from numpy.linalg import cholesky
import matplotlib.pyplot as plt

sampleNo = 10;
mu = 10
sigma = 0.1
y=[]
for i in range(0,100):
    np.random.seed(i)
    s = np.random.normal(mu, sigma, sampleNo )
    sbar = np.mean(s)
    appen = norm.ppf(1-0.05/2)*0.1/np.sqrt(10)
    y.append(sbar)
    print('[%.5f,%.5f]'%((sbar-appen),(sbar+appen)),end='')
    #print( 'The convidence interval is [%f,%f] '%((sbar-appen),(sbar+appen)) )
    
```

```python
#Problem1.2
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(1,100,100)
appen = norm.ppf(1-0.05/2)*0.1/np.sqrt(10)
plt.figure(figsize=(10,6))
for i1 in range(0,100):
    if (abs(y[i1]-10)>appen):
        plt.errorbar(i1+1,y[i1],yerr=appen,fmt='.',ecolor='b',color='b',elinewidth=1,capsize=2)
        print('good')
    else:
        plt.errorbar(i1+1,y[i1],yerr=appen,fmt='.',ecolor='blue',color='b',elinewidth=1,capsize=2)
#plt.hlines(10,0,100)
plt.plot(range(0,100),np.linspace(10,10,100),label='true mean',c='black')
plt.title('Errorbar of CIs') 
plt.xlabel('Sample No.')
plt.ylabel('CI')
plt.legend()
plt.show()
```

```python
#Problem5.2
welding_dataf = pd.read_csv('welding_data(1).csv',header = 0,  index_col = None) # Read .csv file with pandas
# print(welding_dataf.head)
dataf_array = np.asanyarray(welding_dataf) # Convert pandas dataframe as numpy array

dfea1=dataf_array[:,2]
dfea2=dataf_array[:,3]
dfea3=dataf_array[:,4]
dfea4=dataf_array[:,5]
n1=len(dfea1)
time1=range(0,n1)
n2=len(dfea2)
time2=range(0,n2)
n3=len(dfea3)
time3=range(0,n3)
n4=len(dfea4)
time4=range(0,n4)


plt.figure(figsize=(20,16))

plt.subplot(2, 2, 1)
plt.plot(time1,dfea1,c='blue',marker='.',alpha=0.5,label='Configuration 1')
plt.hlines(ucl1,0,n1,linestyle="--")
plt.hlines(lcl1,0,n1,linestyle="--")
plt.title('Control chart for feature 1') 
plt.xlabel('Sample number')
plt.ylabel('Feature characteristic')

plt.subplot(2, 2, 2)
plt.plot(time2,dfea2,c='blue',marker='.',alpha=0.5,label='Configuration 1')
plt.hlines(ucl2,0,n1,linestyle="--")
plt.hlines(lcl2,0,n1,linestyle="--")
plt.title('Control chart for feature 2') 
plt.xlabel('Sample number')
plt.ylabel('Feature characteristic')

plt.subplot(2, 2, 3)
plt.plot(time3,dfea3,c='blue',marker='.',alpha=0.5,label='Configuration 1')
plt.hlines(ucl3,0,n1,linestyle="--")
plt.hlines(lcl3,0,n1,linestyle="--")
plt.title('Control chart for feature 3') 
plt.xlabel('Sample number')
plt.ylabel('Feature characteristic')

plt.subplot(2, 2, 4)
plt.plot(time4,dfea4,c='blue',marker='.',alpha=0.5,label='Configuration 1')
plt.hlines(ucl4,0,n1,linestyle="--")
plt.hlines(lcl4,0,n1,linestyle="--")
plt.title('Control chart for feature 4') 
plt.xlabel('Sample number')
plt.ylabel('Feature characteristic')
```

```python
#Problem5.3
obg_acf_fea1=0;obg_acg_fea1=0
obg_acf_fea2=0;obg_acg_fea2=0
obg_acf_fea3=0;obg_acg_fea3=0
obg_acf_fea4=0;obg_acg_fea4=0
acf=0
for i in range(0,n1):
    if ((dataf_array[i,2]>=lcl1) and (dataf_array[i,2]<=ucl1)):
        if (dataf_array[i,1]==1):
            obg_acf_fea1+=1
        else:
             obg_acg_fea1+=1
    if ((dataf_array[i,3]>=lcl2) and (dataf_array[i,3]<=ucl2)):
        if (dataf_array[i,1]==1):
            obg_acf_fea2+=1
        else:
             obg_acg_fea2+=1
    if ((dataf_array[i,4]>=lcl3) and (dataf_array[i,4]<=ucl3)):
        if (dataf_array[i,1]==1):
            obg_acf_fea3+=1
        else:
             obg_acg_fea3+=1
    if ((dataf_array[i,5]>=lcl4) and (dataf_array[i,5 ]<=ucl4)):
        if (dataf_array[i,1]==1):
            obg_acf_fea4+=1
        else:
             obg_acg_fea4+=1
    if (dataf_array[i,1]==1):
        acf+=1

print('For feature 1, type Ⅰ error rate is %f, type Ⅱ error rate is %f'%((n1-acf-obg_acg_fea1)/(n1-acf),obg_acf_fea1/acf))
print('For feature 2, type Ⅰ error rate is %f, type Ⅱ error rate is %f'%((n1-acf-obg_acg_fea2)/(n1-acf),obg_acf_fea2/acf))
print('For feature 3, type Ⅰ error rate is %f, type Ⅱ error rate is %f'%((n1-acf-obg_acg_fea3)/(n1-acf),obg_acf_fea3/acf))
print('For feature 4, type Ⅰ error rate is %f, type Ⅱ error rate is %f'%((n1-acf-obg_acg_fea4)/(n1-acf),obg_acf_fea4/acf))
```

```python
#Problem5.5
fea1sig=0
fea2sig=0
fea3sig=0
fea4sig=0
obf_acg=0
obf_acf=0
acf=0
for i in range(0,n1):
    if ((dataf_array[i,2]<lcl1) or (dataf_array[i,2]>ucl1)):
        fea1sig=1
    if ((dataf_array[i,3]<lcl2) or (dataf_array[i,3]>ucl2)):
        fea2sig=1
    if ((dataf_array[i,4]<lcl3) or (dataf_array[i,4]>ucl3)):
        fea3sig=1
    if ((dataf_array[i,5]<lcl4) or (dataf_array[i,5]>ucl4)):
        fea4sig=1
    if ((fea1sig+fea2sig+fea3sig+fea4sig)>=3):
        if (dataf_array[i,1]==0):
            obf_acg+=1
        if (dataf_array[i,1]==1):
            obf_acf+=1
    fea1sig=fea2sig=fea3sig=fea4sig=0
    if (dataf_array[i,1]==1):
        acf+=1
print('For rule in (5), type Ⅰ error rate is %f, type Ⅱ error rate is %f'%(obf_acg/(n1-acf),(acf-obf_acf)/acf))
```

```python
#Problem5.6
fea1sig=0
fea2sig=0
fea3sig=0
fea4sig=0
obf_acg=0
obf_acf=0
acf=0
for i in range(0,n1):
    if ((dataf_array[i,2]<lcl1) or (dataf_array[i,2]>ucl1)):
        fea1sig=1
    if ((dataf_array[i,3]<lcl2) or (dataf_array[i,3]>ucl2)):
        fea2sig=1
    if ((dataf_array[i,4]<lcl3) or (dataf_array[i,4]>ucl3)):
        fea3sig=1
    if ((dataf_array[i,5]<lcl4) or (dataf_array[i,5]>ucl4)):
        fea4sig=1
    if ((fea1sig+fea2sig+fea3sig+fea4sig)>=4):
        if (dataf_array[i,1]==0):
            obf_acg+=1
       
        if (dataf_array[i,1]==1):
            obf_acf+=1

    fea1sig=fea2sig=fea3sig=fea4sig=0
            
    if (dataf_array[i,1]==1):
        acf+=1
print('For rule in (6), type Ⅰ error rate is %f, type Ⅱ error rate is %f'%(obf_acg/(n1-acf),(acf-obf_acf)/acf))
```
